# Extração

```python
import pandas as pd
import random
import json
```

```python
lista_carros = pd.read_csv('carros.csv').to_dict(orient='records')
```

```python
for carro in lista_carros:
  carro['info'] = []
```

# Transformação

```python
def adicionar_km(carro):
  km_gerada = random.randint(1000, 50000)
  return km_gerada
```

```python
for carro in lista_carros:
  km_gerada = adicionar_km(carro)
  carro['info'].append({
      'km': km_gerada,
      'status': 'à venda'
  })
```

```python
lista_carros_venda = []
for carro in lista_carros:
  lista_carros_venda.append({
      'carro': carro['carro'],
      'marca': carro['marca'],
      'ano': carro['ano'],
      'info': carro['info']
  })
```

# Carregamento

```python
caminho_arquivo = 'carros_venda.json'

with open(caminho_arquivo, 'w', encoding='utf-8') as f:
    json.dump(lista_carros_venda, f, ensure_ascii=False, indent=4)

print(f"Arquivo {caminho_arquivo} criado com sucesso!")
```
